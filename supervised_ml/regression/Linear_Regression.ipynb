## A.	Explain the main assumptions of Linear Regression in detail.

### Main Assumptions of Linear Regression

### 1. Linearity
- There must be a linear relationship between independent variables (X) and dependent variable (Y).
- The model assumes:  
  Y = β₀ + β₁X₁ + β₂X₂ + ... + ε
- Change in X should cause proportional change in Y.
- If violated → Model gives biased or inaccurate predictions.

---

### 2. Independence of Errors
- Residuals (errors) must be independent of each other.
- One observation’s error should not influence another.
- Commonly violated in time series data.
- If violated → Standard errors and significance tests become unreliable.

---

### 3. Homoscedasticity (Constant Variance)
- The variance of residuals should remain constant across all levels of X.
- The spread of errors should be uniform.
- If variance changes → It is called heteroscedasticity.
- If violated → Confidence intervals and hypothesis tests become invalid.

---

### 4. Normality of Errors
- Residuals should be normally distributed.
- Assumed as: ε ~ N(0, σ²)
- Important for hypothesis testing and confidence intervals.
- If violated → Predictions may work, but statistical inference becomes weak.

---

### 5. No Multicollinearity
- Independent variables should not be highly correlated with each other.
- High correlation among predictors causes instability in coefficients.
- If violated → Large standard errors and unreliable coefficient estimates.

---

### 6. No Autocorrelation
- Residuals should not show patterns over time.
- Important especially in time series data.
- If violated → Model assumptions break and results become unreliable.

### B.	What is the difference between R-squared and Adjusted R-squared?

### Difference Between R-squared and Adjusted R-squared

### 1. R-squared (Coefficient of Determination)

- Measures how much variation in the dependent variable is explained by independent variables.
- Value ranges between 0 and 1.
- Formula:

  R² = 1 − (SS_res / SS_total)

- Higher R² means better model fit.
- Problem: R² always increases when more variables are added, even if they are irrelevant.

---

### 2. Adjusted R-squared

- Modified version of R² that adjusts for number of predictors in the model.
- Penalizes the addition of unnecessary variables.
- Formula:

  Adjusted R² = 1 − [(1 − R²)(n − 1) / (n − k − 1)]

  where:
  n = number of observations  
  k = number of independent variables  

- Can increase or decrease when new variables are added.
- Gives a more reliable measure for multiple regression.

---

### Key Differences

| Basis | R-squared | Adjusted R-squared |
|--------|------------|--------------------|
| Meaning | Explains total variance | Explains variance adjusted for predictors |
| Effect of adding variables | Always increases | Increases only if variable is useful |
| Penalizes complexity | No | Yes |
| Better for multiple regression | No | Yes |

---

### Conclusion

- R² shows goodness of fit.
- Adjusted R² shows goodness of fit considering model complexity.
- For multiple linear regression, Adjusted R² is preferred.

## C.	What are the different types of Regularization techniques in Regression. Explain in detail with cost functions of each technique.

### Regularization Techniques in Regression

### Introduction

Regularization techniques are used to prevent overfitting in regression models by adding a penalty term to the cost function.  
They reduce model complexity and improve generalization.

---

### 1. Ridge Regression (L2 Regularization)

### Concept
- Adds a penalty equal to the square of the magnitude of coefficients.
- Shrinks coefficients toward zero.
- Does NOT make coefficients exactly zero.

### Cost Function

Ordinary Least Squares (OLS):

J(β) = Σ (yᵢ − ŷᵢ)²

Ridge Regression:

J(β) = Σ (yᵢ − ŷᵢ)² + λ Σ βⱼ²

Where:
- λ = regularization parameter
- βⱼ = model coefficients

### Key Points
- Controls multicollinearity.
- Keeps all features.
- Larger λ → stronger shrinkage.
- Best when all features are important.

---

### 2. Lasso Regression (L1 Regularization)

### Concept
- Adds penalty equal to absolute value of coefficients.
- Can shrink some coefficients exactly to zero.
- Performs feature selection automatically.

### Cost Function

J(β) = Σ (yᵢ − ŷᵢ)² + λ Σ |βⱼ|

Where:
- λ controls strength of penalty.

## Key Points
- Produces sparse models.
- Eliminates irrelevant features.
- Useful when many features exist.

---

###3. Elastic Net Regression

### Concept
- Combination of Ridge (L2) and Lasso (L1).
- Used when dataset has many correlated features.

### Cost Function

J(β) = Σ (yᵢ − ŷᵢ)² 
       + λ₁ Σ |βⱼ| 
       + λ₂ Σ βⱼ²

OR

J(β) = Σ (yᵢ − ŷᵢ)² + λ [ α Σ |βⱼ| + (1 − α) Σ βⱼ² ]

Where:
- λ = overall regularization strength
- α = mixing parameter (0 ≤ α ≤ 1)
  - α = 1 → Lasso
  - α = 0 → Ridge

### Key Points
- Handles multicollinearity.
- Performs feature selection.
- More flexible than Ridge or Lasso alone.

---

### Comparison of Regularization Techniques

| Feature | Ridge | Lasso | Elastic Net |
|----------|--------|--------|--------------|
| Penalty Type | L2 (β²) | L1 (|β|) | L1 + L2 |
| Feature Selection | No | Yes | Yes |
| Handles Multicollinearity | Yes | Moderate | Yes |
| Coefficients Become Zero | No | Yes | Yes |

---

### Role of Regularization Parameter (λ)

- λ = 0 → Equivalent to simple Linear Regression.
- Large λ → High bias, Low variance.
- Small λ → Low bias, High variance.

---

### Conclusion

- Ridge → Best when all features matter.
- Lasso → Best for feature selection.
- Elastic Net → Best when features are correlated and large in number.
- Regularization improves model generalization and prevents overfitting.

## D.	How logistic regression works for multiclass classification. Explain in detail.

### Logistic Regression for Multiclass Classification

### Introduction

Logistic Regression is primarily used for binary classification.
However, it can be extended to handle multiclass classification problems using two main approaches:

1. One-vs-Rest (OvR)
2. Multinomial Logistic Regression (Softmax Regression)

---

### 1. Binary Logistic Regression (Basic Idea)

For binary classification:

P(Y = 1 | X) = 1 / (1 + e^-(β₀ + β₁X₁ + ... + βₙXₙ))

This uses the Sigmoid function to convert linear output into probability between 0 and 1.

---

### 2. One-vs-Rest (OvR) / One-vs-All Approach

### Concept

- For K classes, build K separate binary classifiers.
- Each classifier distinguishes one class vs all other classes.

Example:
If classes = {A, B, C}

We build:
- Classifier 1: A vs (B, C)
- Classifier 2: B vs (A, C)
- Classifier 3: C vs (A, B)

### How Prediction Works

- Each classifier gives a probability score.
- The class with the highest probability is selected.

### Mathematical Form

For each class k:

P(Y = k | X) = sigmoid(β_k^T X)

Final prediction:

Predicted Class = argmax P(Y = k | X)

### Advantages
- Simple and easy to implement.
- Works well when classes are well separated.

### Disadvantages
- Classifiers are trained independently.
- Probability outputs may not sum to 1.

---

### 3. Multinomial Logistic Regression (Softmax Regression)

### Concept

Instead of building multiple binary classifiers, a single model is trained that directly predicts probabilities for all classes simultaneously.

### Softmax Function

For K classes:

P(Y = k | X) = e^(β_k^T X) / Σ e^(β_j^T X)

Where:
- β_k = coefficient vector for class k
- Denominator = sum over all classes

### Properties

- Probabilities sum to 1.
- Directly models multiple classes.

### Cost Function (Cross-Entropy Loss)

J(β) = - Σ Σ yᵢk log(Pᵢk)

Where:
- yᵢk = 1 if observation i belongs to class k, else 0
- Pᵢk = predicted probability

---

### Difference Between OvR and Multinomial

| Feature | One-vs-Rest | Multinomial (Softmax) |
|----------|--------------|------------------------|
| Number of Models | K models | Single model |
| Probability Sum | May not sum to 1 | Always sums to 1 |
| Training | Separate classifiers | Joint optimization |
| Preferred When | Small datasets | Large datasets, well-separated classes |

---

### Example

Suppose we classify houses into:

1. Low Price
2. Medium Price
3. High Price

Using Softmax:
The model calculates 3 probability scores:

P(Low), P(Medium), P(High)

The highest probability becomes the predicted class.

---

### Summary

- Logistic Regression can handle multiclass problems.
- Two main approaches:
  - One-vs-Rest (multiple binary models)
  - Multinomial (Softmax regression)
- Softmax is mathematically stronger and commonly preferred.
- Uses Cross-Entropy loss for optimization.

## E.	Explain the performance metrics of logistic regression.

### Performance Metrics of Logistic Regression

Logistic Regression is used for classification problems. 
Its performance is evaluated using classification metrics rather than regression metrics.

---

### 1. Confusion Matrix

A table used to evaluate classification performance.

|                | Predicted Positive | Predicted Negative |
|---------------|-------------------|-------------------|
| Actual Positive | True Positive (TP) | False Negative (FN) |
| Actual Negative | False Positive (FP) | True Negative (TN) |

It forms the base for calculating all other metrics.

---

### 2. Accuracy

### Definition:
Proportion of correctly predicted observations.

Accuracy = (TP + TN) / (TP + TN + FP + FN)

### When to Use:
- When dataset is balanced.

### Limitation:
- Not reliable for imbalanced datasets.

---

### 3. Precision

### Definition:
Proportion of correctly predicted positive observations out of total predicted positives.

Precision = TP / (TP + FP)

### Meaning:
- How many predicted positives are actually correct.

### Important When:
- False Positives are costly (e.g., spam detection).

---

### 4. Recall (Sensitivity / True Positive Rate)

### Definition:
Proportion of correctly predicted positives out of actual positives.

Recall = TP / (TP + FN)

### Meaning:
- How many actual positives were correctly identified.

### Important When:
- False Negatives are costly (e.g., medical diagnosis).

---

### 5. F1-Score

### Definition:
Harmonic mean of Precision and Recall.

F1 = 2 × (Precision × Recall) / (Precision + Recall)

### Meaning:
- Balances Precision and Recall.
- Useful for imbalanced datasets.

---

### 6. Specificity (True Negative Rate)

Specificity = TN / (TN + FP)

Measures how well the model identifies negative cases.

---

### 7. ROC Curve (Receiver Operating Characteristic)

- Graph between:
  - True Positive Rate (Recall)
  - False Positive Rate (FPR)

FPR = FP / (FP + TN)

Shows model performance at different threshold values.

---

### 8. AUC (Area Under Curve)

- Area under ROC curve.
- Value ranges between 0 and 1.

Interpretation:
- 0.5 → Random model
- 1.0 → Perfect model
- Higher AUC → Better classification ability

---

### 9. Log Loss (Cross-Entropy Loss)

Used to measure probability error.

Log Loss = - [ y log(p) + (1 - y) log(1 - p) ]

Lower Log Loss → Better model.

Penalizes incorrect confident predictions heavily.

---

### Summary Table

| Metric | Best Used When | Focus |
|--------|---------------|-------|
| Accuracy | Balanced dataset | Overall correctness |
| Precision | False positives costly | Prediction reliability |
| Recall | False negatives costly | Detection ability |
| F1-score | Imbalanced dataset | Balance of precision & recall |
| AUC-ROC | Comparing models | Overall discrimination power |
| Log Loss | Probabilistic models | Confidence quality |

---

### Conclusion

- Logistic Regression performance depends on problem type.
- For balanced data → Accuracy works.
- For imbalanced data → F1-score and AUC are preferred.
- For probability evaluation → Log Loss is important.
